```python
# –ø—Ä–æ–≤–µ—Ä–∫–∞ –≥–∏–ø–æ—Ç–µ–∑—ã –æ —Ä–∞–≤–µ–Ω—Å—Ç–≤–µ —Å—Ä–µ–¥–Ω–µ–≥–æ –¥–æ—Ö–æ–¥–∞
import scipy.stats as st
# –≤—ã–±–∏—Ä–∞–µ–º —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ 5%
alpha = 0.05
```

```python
# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å —É—à–µ–¥—à–∏–º–∏ –∫–ª–∏–µ–Ω—Ç–∞–º–∏
churn_df = data.query('churn == 1')

# –¥–∞—Ç–∞—Ñ—Ä–µ–π–º —Å –æ—Å—Ç–∞–≤—à–∏–º–∏—Å—è –∫–ª–∏–µ–Ω—Ç–∞–º–∏
loyal_df = data.query('churn != 1')
#loyal_df
```

```python
# —Å—Ä–µ–¥–Ω–Ω–∏–π –æ—Ü–µ–Ω–æ—á–Ω—ã–π –¥–æ—Ö–æ–¥ —É—à–µ–¥—à–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤
churn_value = churn_df['est_salary']
display(f'–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ü–µ–Ω–æ—á–Ω–æ–≥–æ –¥–æ—Ö–æ–¥–∞ —É—à–µ–¥—à–∏—Ö –∫–ª–∏–µ–Ω—Ç–æ–≤: {churn_df.est_salary.mean().round(2)}')

# —Å—Ä–µ–¥–Ω–Ω–∏–π –æ—Ü–µ–Ω–æ—á–Ω—ã–π –¥–æ—Ö–æ–¥ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∫–ª–∏–µ–Ω—Ç–æ–≤
loyal_value = loyal_df['est_salary']
display(f'–°—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –æ—Ü–µ–Ω–æ—á–Ω–æ–≥–æ –¥–æ—Ö–æ–¥–∞ –æ—Å—Ç–∞–≤—à–∏—Ö—Å—è –∫–ª–∏–µ–Ω—Ç–æ–≤: {loyal_df.est_salary.mean().round(2)}')

results = st.ttest_ind(churn_value, loyal_value, equal_var=False)

display(f'p-value: {results.pvalue}')

# –ø—Ä–æ–≤–µ—Ä—è–µ–º p-value
if results.pvalue < alpha:
    display('–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É: —Å—Ä–µ–¥–Ω–∏–µ –æ—Ü–µ–Ω–æ—á–Ω—ã–µ –¥–æ—Ö–æ–¥—ã –≥—Ä—É–ø–ø –Ω–µ –æ–¥–∏–Ω–∞–∫–æ–≤—ã')
else:
    display('–ù–µ—Ç –æ—Å–Ω–æ–≤–∞–Ω–∏–π –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É')
```
